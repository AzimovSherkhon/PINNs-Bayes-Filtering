```python
import numpy as np
all_predictions = np.load("pinn_bayes_solutions_0_50_k_0.9_2.npy")
N = 401
L = 2
dt = 0.01
all_predictions10 = all_predictions[:,:1000]
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-L, L, N)
t = np.linspace(0, dt * all_predictions10.shape[1], all_predictions10.shape[1])
X, T = np.meshgrid(x, t, indexing="ij")

fig = plt.figure(figsize=(16, 12))
ax = fig.add_subplot(111, projection="3d")

surf = ax.plot_surface(
    X, T, all_predictions10,
    cmap="cividis",
    edgecolor="none",
    rstride=1, cstride=1,
    alpha=0.9,
    linewidth=0,
    antialiased=True
)

ax.set_xlabel("System State ($x$)", fontsize=14, labelpad=15)
ax.set_ylabel("Time ($t$)", fontsize=14, labelpad=15)
ax.set_zlabel("Probability Density", fontsize=14, labelpad=15)
ax.set_title("PINNs Bayes $[0, 10]$", fontsize=16, fontweight="bold", pad=15)

colorbar = fig.colorbar(surf, ax=ax, shrink=0.6, aspect=12, pad=0.1)
colorbar.set_label("PDF Value", fontsize=12, labelpad=10)
colorbar.ax.tick_params(labelsize=10)

ax.tick_params(axis="both", which="major", labelsize=10)
ax.view_init(elev=30, azim=240)

plt.tight_layout()
plt.savefig("3D_PDF_Evolution_0-10_k=0.9_tobs=2.png", dpi=300, bbox_inches="tight", format="png")
plt.show()
plt.close(fig)
```
