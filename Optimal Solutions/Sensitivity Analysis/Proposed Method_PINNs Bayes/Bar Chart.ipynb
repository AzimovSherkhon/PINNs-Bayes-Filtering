```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_palette("colorblind")

plt.rcParams.update({
    "font.size": 18,  
    "axes.titlesize": 22,  
    "axes.labelsize": 20,  
    "xtick.labelsize": 12,  
    "ytick.labelsize": 12,  
    "legend.fontsize": 18,  
})

ensemble_particle_counts = [r'$10^2$', r'$10^2$', r'$10^4$', r'$10^4$']
obs_frequencies = [1.0, 2.0, 1.0, 2.0]

mae_enkf_05 = [0.0613, 0.1069, 0.0388, 0.0698]
mae_pf_05 = [0.1186, 0.2207, 0.0140, 0.0032]

mae_enkf_09 = [0.0402, 0.0577, 0.0234, 0.0375]
mae_pf_09 = [0.0331, 0.0398, 0.0046, 0.0054]

bar_width = 0.4
x = np.arange(len(ensemble_particle_counts))  # X locations for the groups

fig, axes = plt.subplots(1, 2, figsize=(24, 12), sharey=True)

colors = sns.color_palette()

axes[0].bar(x - bar_width/2, mae_enkf_05, bar_width, label='MAE (EnKF)', color=colors[0])
axes[0].bar(x + bar_width/2, mae_pf_05, bar_width, label='MAE (PF)', color=colors[1])
axes[0].set_xticks(x)
axes[0].set_xticklabels([f'Ensemble/Particle: {ensemble_particle_counts[i]}\nObservation Frequency: {obs_frequencies[i]}' for i in range(len(ensemble_particle_counts))])
axes[0].set_title(r'Results for $\kappa=0.5$ (Low Noise)')
axes[0].set_ylabel('MAE')
axes[0].legend()

axes[1].bar(x - bar_width/2, mae_enkf_09, bar_width, label='MAE (EnKF)', color=colors[0])
axes[1].bar(x + bar_width/2, mae_pf_09, bar_width, label='MAE (PF)', color=colors[1])
axes[1].set_xticks(x)
axes[1].set_xticklabels([f'Ensemble/Particle: {ensemble_particle_counts[i]}\nObservation Frequency: {obs_frequencies[i]}' for i in range(len(ensemble_particle_counts))])
axes[1].set_title(r'Results for $\kappa=0.9$ (High Noise)')
axes[1].legend()

plt.tight_layout()
plt.savefig("mae_results_EnKF_PF.png", dpi=300)
plt.show()
```
